# Scaling

``ess.nmx`` provides scaling methods as well as ``mtz`` io.

```python
import sciline as sl
import os

from ess.nmx.mtz_io import (
    read_mtz_file,
    FileName,
    MTZFilepath,
)
from ess.nmx.scaling import (
    reduce_mtz,
    merge_mtz_dataframes,
    get_reciprocal_asu,
    get_space_group,
    MergedMtzDataFrame,    
    SpaceGroupDesc,
)

workflow = sl.Pipeline(providers=[
    read_mtz_file,
    reduce_mtz,
    merge_mtz_dataframes,
    get_reciprocal_asu,
    get_space_group
],
    params = {
        SpaceGroupDesc: 'C 1 2 1'  # Replace with the correct space group if needed
    }
)

dir_path = '/home/sunyoung/ESS/pyscale/assets/test_large/'
file_path_table = sl.ParamTable(
    row_dim=FileName,
    columns={
        MTZFilepath: [dir_path + file_name for file_name in os.listdir(dir_path)] # [:20]
    }
)

workflow.set_param_table(file_path_table)
# workflow.visualize(MergedMtzDataFrame, graph_attr={"rankdir": "LR"})
```

```python
df = workflow.compute(MergedMtzDataFrame)
df
```
